```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("bank-additional-full.csv",sep=';')
data
```

```python
data= data.drop(columns = ['job'])
data= data.drop(columns = ['marital'])
data= data.drop(columns = ['education'])
data= data.drop(columns = ['default'])
data= data.drop(columns = ['housing'])
data= data.drop(columns = ['loan'])
data= data.drop(columns = ['contact'])
data= data.drop(columns = ['month'])
data= data.drop(columns = ['day_of_week'])
data= data.drop(columns = ['poutcome'])
data= data.drop(columns = ['y'])
```

```python
X = data.drop(columns = ["euribor3m"])
y = data["euribor3m"]
```

```python
X.dtypes
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=15 )
```

```python
from fedot.api.main import Fedot

model = Fedot(problem='regression', seed=5, timeout=1, show_progress=True, max_pipeline_fit_time=1)
model.fit(features=X_train, target=y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

print(f"Оценка качества модели по метрике MAE: {mean_absolute_error(y_test, model.predict(X_test))}")
print(f"Оценка качества модели по метрике MSE: {mean_squared_error(y_test, model.predict(X_test))}")
```

```python
plt.scatter(model.predict(X_test), y_test)
plt.show()
```

#### Построение собственной модели

```python
X = data.drop(columns = ['euribor3m'])
y = data['euribor3m']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=15 )
```

#### Как модель регрессии возьмем, случайное дерево

```python
# Fitting the Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor_rf.fit(X_train, y_train)
```

#### Оценка модели с помощью метрик MAE и MSE

```python
print(f"Оценка качества модели по метрике MAE: {mean_absolute_error(y_test, regressor_rf.predict(X_test))}")
print(f"Оценка качества модели по метрике MSE: {mean_squared_error(y_test, regressor_rf.predict(X_test))}")
```

#### Scatter для модели построенной мной

```python
plt.scatter(model.predict(X_test), y_test)
plt.show()
```
