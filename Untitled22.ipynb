```python
from scipy.optimize import minimize #Импорт функции minimize из модуля scipy.optimize, которая используется для минимизации функций.

# Определение функции
def f(x):
    x1, x2 = x
    return 45*x1**2-88*x1*x2+45*x2**2+102*x1**2+268*x2-21

# Определение градиента функции
def grad_f(x):#Определение функции grad_f(x), которая вычисляет градиент функции для заданных x1 и x2. 
    #В данном случае, так как используется метод CG (Conjugate Gradient), градиент передается как список.
    x1, x2 = x
    return [45*x1**2-88*x1*x2+45*x2**2+102*x1**2+268*x2-21]

# Задание начальной точки
x0 = [-1,1]

# Минимизация функции с использованием метода наискорейшего спуска
result = minimize(f, x0, method='CG', jac=grad_f)#Вызов функции minimize для поиска минимума функции f с использованием метода CG (Conjugate Gradient) и заданным градиентом. 
#Результат сохраняется в переменной result.

# Вывод результатов
print('Точка минимума:', result.x)
print('Значение функции в точке минимума:', result.fun)
```

```python
# Определение функции f
def f(x):
    x1, x2 = x[0], x[1]
    return 45*x1**2-88*x1*x2+45*x2**2+102*x1**2+268*x2-21

# Определение градиента функции f
def grad_f(x):
    x1, x2 = x[0], x[1]
    df_dx1 = 90 * x1 - 88 * x2 + 204 * x1
    df_dx2 = -88 * x1 + 90 * x2 + 268
    return np.array([df_dx1, df_dx2])


# Определение функции для метода наискорейшего спуска
def gradient_descent(f, grad_f, x0, learning_rate=0.01, epsilon=1e-6, max_iterations=1000):
    x = x0
    iteration = 0
    while np.linalg.norm(grad_f(x)) > epsilon and iteration < max_iterations:
        gradient = grad_f(x)
        x = x - learning_rate * gradient
        iteration += 1
    return x

# Начальная точка
x0 = np.array([0, 1])

# Применение метода наискорейшего спуска
minimum_point = gradient_descent(f, grad_f, x0, learning_rate=0.001, epsilon=1e-8, max_iterations=10000)

# Вывод результатов
minimum_value = f(minimum_point)
print("Точка минимума:", minimum_point)
print("Значение функции в точке минимума:", minimum_value)
```

## Задние 4

```python
#Определение функции f
def f(x):
    x1, x2 = x
    return 45*x1**2-88*x1*x2+45*x2**2+102*x1**2+268*x2-21

#Определение градиента функции f
def grad_f(x):
    x1, x2 = x[0], x[1]
    df_dx1 = 90 * x1 - 88 * x2 + 204 * x1
    df_dx2 = -88 * x1 + 90 * x2 + 268
    return np.array([df_dx1, df_dx2])

#Определение функции для метода наискорейшего спуска
def gradient_descent(f, grad_f, x0, learning_rate=0.01, epsilon=1e-6, max_iterations=1000):
    x = x0
    iteration = 0
    iterations = []
    while np.linalg.norm(grad_f(x)) > epsilon and iteration < max_iterations:
        gradient = grad_f(x)
        x = x - learning_rate * gradient
        iterations.append(iteration)
        iteration += 1
    return x, iterations

#Начальная точка
x0 = np.array([-1, 1])

#Применение метода наискорейшего спуска
minimum_point, iteration_history = gradient_descent(f, grad_f, x0, learning_rate=0.001, epsilon=1e-8, max_iterations=10000)

#Вывод результатов
minimum_value = f(minimum_point)
print("Точка минимума:", minimum_point)
print("Значение функции в точке минимума:", minimum_value)

#Визуализация количества итераций
plt.plot(iteration_history)
plt.xlabel('Iteration')
plt.ylabel('Function Value')
plt.title('Convergence of Gradient Descent')
plt.show()
```
