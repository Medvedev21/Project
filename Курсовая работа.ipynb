Cгрупировал столбецы месяцам и столбецы количества рейсов

# 1.1

```python
number_flights = flights['day'].groupby(flights['month']).count()
number_flights = number_flights.reset_index()
number_flights.rename(columns={'day': 'flights'}, inplace=True)
number_flights.replace({1 : 'January', 2 : 'February', 3 : 'March', 4 : 'April', 5 : 'May', 6 : 'June', 7 : 'July', 8 : 'August', 9 : 'September', 10 : 'October', 11 : 'November', 12 : 'December' }, inplace = True)
number_flights
```

Вывожу диаграмму, результат хороший 

```python
x = number_flights['month']
y = number_flights['flights']

fig, ax = plt.subplots()

ax.bar(x, y)

ax.set_xlabel('Месяцы',fontsize = 15)
ax.set_ylabel('Количество рейсов',fontsize = 15)

fig.set_figwidth(20)    #  ширина Figure
fig.set_figheight(10)    #  высота Figure

plt.show()
```

Количество рейсов больше всего пришлось на 7 месяц, а спад пришел на 2 месяц

# 1.2

делаю групировку по дням недели с количеством рейсов

```python
flights_day = flights['day'].groupby(flights['day_of_week']).count()
flights_day = flights_day.reset_index()
flights_day.rename(columns={'day': 'flights'}, inplace=True)
flights_day
```

Вывожу диаграмму, результат хороший 

```python
x = flights_day['day_of_week']
y = flights_day['flights']

fig, ax = plt.subplots()

ax.bar(x, y)

ax.set_xlabel('Дни недели',fontsize = 15)
ax.set_ylabel('Количество рейсов',fontsize = 15)

fig.set_figwidth(20)    #  ширина Figure
fig.set_figheight(10)    #  высота Figure

plt.show()
```

Количество рейсов распределено по рабочим дням, а в выходные виден спад.

```python
from_city = flights['origin_airport'].value_counts()
to_city = flights['destination_airport'].value_counts()
summa_flights = from_city.append(to_city)
summa_flights
```

```python
group_sum = summa_flights.groupby(summa_flights.index).sum()
airports_group = pd.DataFrame(group_sum.sort_values(ascending = False), columns = ['flights'])
airports_group
```

```python
group_city = airports.set_index('iata_code').join(airports_group)
group_city
```

```python
tp10_city = group_city['flights'].groupby(group_city['city']).sum().sort_values(ascending = False).head(10)
tp10_city = tp10_city.reset_index()
tp10_city
```
